```python
import pandas as pd

# Load the PERFECT Spotify dataset directly from GitHub (160k tracks, 1921–2020)
# Columns: artist, track, popularity, danceability, energy, valence (happiness), acousticness, tempo, etc.
url = "https://raw.githubusercontent.com/gabminamedez/spotify-data/master/data.csv"
df = pd.read_csv(url)

# Quick check: No errors, just glory
print("Dataset loaded! Shape:", df.shape)  # → (170313, 19) or similar
print("\nFirst 5 tracks:")
print(df[['name', 'id', 'valence', 'danceability']].head())

# Create your "mood score" (simple formula: happy * danceable = overall vibe)
df['mood'] = df['valence'] * df['danceability']
print("\nSample moods:")
print(df[['name', 'id', 'mood']].head())
```

```python
import matplotlib.pyplot as plt

# Simple bar: Top 5 happiest tracks by mood score
top_moods = df.nlargest(5, 'mood')[['name', 'id', 'mood']]
plt.figure(figsize=(10, 6))
plt.barh(range(len(top_moods)), top_moods['mood'])
plt.yticks(range(len(top_moods)), top_moods['name'] + ' (' + top_moods['id'] + ')')
plt.xlabel('Mood Score')
plt.title('Top 5 Happiest Tracks (Day 1 Discovery)')
plt.savefig('day1_top_moods.png')  # Saves image in Colab
plt.show()
print("Image saved! Ready for GitHub.")
```
